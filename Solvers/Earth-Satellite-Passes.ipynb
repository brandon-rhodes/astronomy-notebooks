```python
from scipy import optimize
from skyfield.api import JulianDate, earth

tle = """TIANGONG 1              
1 37820U 11053A   14314.79851609  .00064249  00000-0  44961-3 0  5637
2 37820  42.7687 147.7173 0010686 283.6368 148.1694 15.73279710179072
"""

sat = earth.satellite(tle)
```

The Skyfield API makes it easy, given a particular date and time, to observe a satellite. We start by asking an Earth location “What were your coordinates at this date and time?” and then ask that location to “observe” the satellite and tell us what its relative direction was. We could ask for any of several kinds of coordinate. Because we are interested in whether the satellite is above the horizon or not, we will ask for altazimuth coordinates.

To make things simple, we will ignore the azimuth and distance to the satellite for now and write a function that takes a time in TAI atomic time and returns the satellite’s altitude. Because these are both smoothly varying numbers, we will be able to submit this function to a SciPy solver for analysis — from a math point of view, this is a function that takes a number in and produces a number back out:

```python
boston = earth.topos('42.3581 N', '71.0636 W')

def f(t):
    alt, az, distance = boston(tai=t).observe(sat).altaz()
    return alt.degrees
```

Computing a satellite position is fairly expensive with the current SGP4 implementation that Skyfield uses because of Skyfield’s design: it does everything using only Python and NumPy so that it can be used in environments like the Google App Engine. So it takes a long time to, for example, ask a satellite where it was every minute for a whole day. But let’s do so — just once — so we have a complete, detailed picture of how satellite altitude behaves:

```python
t0 = JulianDate(utc=(2014, 11, 10)).tai
one_minute = 1.0 / 24.0 / 60.0
one_second = one_minute / 60.0
every_minute = arange(0, 1.0, one_minute)
t = t0 + every_minute

dt_smooth = JulianDate(tai=t).utc_datetime()
altitude_smooth = f(t)
plot(dt_smooth, altitude_smooth)
```

The above graph makes perfect sense. A satellite orbit can be thought of as an ellipse that hangs stable in space while the Earth rotates beneath it. In this graph, we can see that for part of the day Boston is quite distant from the (rough) great circle that the satellite travels above the planet’s surface, so through those hours the satellite stays well below the horizon. But then Boston rotates around and beneath the arc of the orbit, so during those hours the satellite passes repeatedly overhead and its altitude goes briefly positive.

What we would like to do is locate each point where the satellite reaches a maximum altitude — but to do so quickly, without taking the full 15 seconds that my laptop required to produce the every-minute-of-the-day graph above.

How much resolution do we need — how many points per orbit — to be able to identify at least one point that is over on Boston’s side of the Earth and where we could start a search? If for each orbit of the satellite we ask for six points, then they should form a rough hexagon in space around the Earth — and one of those points should have a clearly higher altitude than its neighbors, and thus draw our attention to the part of the orbit we need more information about.

That is, a six-points-per-revolution sampling of the satellite position should preserve enough of the shape of the above graph that we can see each of the maxima — even though we will no longer yet know exactly how high it was.

To learn this we need to look at the orbital element that tells us how fast the satellite moves. I should think about adding a better name for it to Skyfield:

```python
orbit_period_per_minute = tau / sat._sgp4_satellite.no
orbit_period = orbit_period_per_minute / 24.0 / 60.0
print 1.0 / orbit_period, 'revolutions per day'
```

This agrees fairly well with the same number as stated in sources like the Wikipedia, so we appear to have computed it correctly!

We can now ask for the satellite position over a whole day, but with only six points per complete orbit, and the graph should come back quite quickly:

```python
step = orbit_period / 6.0
t = t0 + arange(0, 1.0, step)
dt = JulianDate(tai=t).utc_datetime()
altitude = f(t)
plot(dt, altitude)
```

That was much quicker!

And from the shape of the graph, it looks like we were right. With only six points per orbit we are unlikly to happen to find the exact moments where the satellite is highest — but we have at least identified the rough time of each peak and can start searching there!

Let us use the magic of NumPy to ask for the peaks. They are the places where the altitude stops getting bigger and starts getting smaller:

```python
left_diff = ediff1d(altitude, to_begin=0.0)
right_diff = ediff1d(altitude, to_end=0.0)
maxima = (left_diff > 0.0) & (right_diff < 0.0)

plot(dt, altitude)
plot(dt[maxima], altitude[maxima], 'go')
```

Every one of these red points is a candidate from which we can search for a maximum — and we can limit the search because we know that the maximum should be within a single `step` to either side of the point.

One peculiarity of SciPy extrama solvers is that they are all minimizers — so we need a function that returns the negative altitude that a minimizer can work on:

```python
def minusf(t):
    return -f(t)
```

With this function in hand, we can set one of the standard SciPy solvers to work on finding the minimum that lies close to each of the green dots in the above plot — each of which should be a maximum altitude reached by the satellite during one of its passes overhead (if the resulting value is greater than zero, of course).

```python
def find_highest(t):
    result = optimize.minimize_scalar(
        minusf, bracket=[t+step,t,t-step], tol=one_second / t)
    return result.x
```

```python
t_highest = [find_highest(ti) for ti in t[maxima]]
dt_highest = JulianDate(tai=t_highest).utc_datetime()

plot(dt, altitude)
plot(dt[maxima], f(t[maxima]), 'go')
plot(dt_highest, f(t_highest), 'ro')
```

Has this approach really made a successful leap from the limited information provided by six points per orbit and its resulting green dots to a real list of maximum altitudes? We can compare these red dots with the very expensive curve we computed earlier that measured the altitude every minute:

```python
plot(dt_smooth, altitude_smooth)
plot(dt_highest, f(t_highest), 'ro')
```

Because we set the solver to find the moments of maximum altitude to a time resolution of one second (see the `tol` argument provided to the solver), we have in fact found even better maximum altitudes than were possible with the more expensive curve.

# Risings and Settings

The SciPy documentation recommends the `brentq()` function as their best and most robust algorithm for finding where a curve crosses zero — which is the calculation we need if we are to determine when a satellite crests the horizon, or when it again descends beneath it.

But for `brentq()` to work it needs to be kicked off by being given both a point at which the target function is *less* than zero and also a point at which it is *more* than zero. Therefore we need to work starting from a list of points, like the maxima given above, that all represent moments when the satellite is above the horizon.

Where should we place the “other point” for each calculation, that must be below the horizon? One simple choice is to go one-third of the way back or forward around the satellite’s orbit, since that should be more than enough to take it below the horizon (assuming that the satellite is not geocentric, or in a very slow-moving and distant orbit). Since our `step` is intended to represent $1\over 6$ of an orbit, we can simply take twice the `step` as being roughly $1\over 3$.

This leads us to an approach like:

```python
def find_rising(t):
    """Provide a moment of maximum altitude as `t`."""
    return optimize.brentq(f, t - 2*step, t)

def find_setting(t):
    """Provide a moment of maximum altitude as `t`."""
    return optimize.brentq(f, t + 2*step, t)
```

```python
passes = [ti for ti in t_highest if f(ti) > 0.0]
dt_passes = JulianDate(tai=passes).utc_datetime()

t_rising = [find_rising(ti) for ti in passes]
dt_rising = JulianDate(tai=t_rising).utc_datetime()

t_setting = [find_setting(ti) for ti in passes]
dt_setting = JulianDate(tai=t_setting).utc_datetime()

plot(dt_smooth, altitude_smooth)
plot(dt_highest, f(t_highest), 'ro')
plot(dt_rising, f(t_rising), 'go')
plot(dt_setting, f(t_setting), 'bo')
xlim(dt_rising[0], dt_setting[-1])
```

The above algorithm indeed seems to quickly find both the rising and setting that are immediately adjacent to these moments of highest altitude.

By printing all three values, we can construct a table of satellite passes.

```python
for rising, highest, setting in zip(dt_rising, dt_passes, dt_setting):
    print rising.strftime('%Y-%m-%d %H:%M:%S '),
    print highest.strftime('%Y-%m-%d %H:%M:%S '),
    print setting.strftime('%Y-%m-%d %H:%M:%S')
```
